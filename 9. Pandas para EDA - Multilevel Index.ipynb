## Indexing multinivel

```python
dfm.head(3)
```

Usar loc funciona igual, cuando sólo se considera el nivel más externo.

```python
dfm.loc['2014-01-02']
```

iloc es agnóstico al multiíndice. Se fija en el nivel más interno solamente.

```python
dfm.iloc[3,3]
```

Se pueden pasar múltiples argumentos a loc para accesar dos instancias **específicas** de los índices.

```python
dfm.loc['2014-01-02','GOOGL']
```

La forma correcta de hacer esto es pasando una tupla.

```python
dfm.loc[('2014-01-02','GOOGL')]
```

Puede pasarse una tupla para index y un string para column.

```python
dfm.loc[('2014-01-02','GOOGL'),'low']
```

## Slicing multinivel

Seleccionar múltiples días específicos

```python
dfm.loc[(['2014-01-06','2015-01-06','2015-02-06'],slice(None)),:]
```

Se pueden pasar dos listas dentro de una tupla para ver instancias de varios niveles.

```python
dfm.loc[(['2014-01-06','2015-01-07','2015-02-10'],['MSFT','AMZN']),:]
```

Slicing a través de nivel exterior y columna

```python
dfm.loc['2014-01-06':'2014-03-06','open':'high']
```

Mezclando niveles

```python
dfm.loc[(slice('2014-01-06','2014-03-06'),['GOOGL','FB']),'open':'high']
```

Seleccionando nivel interior

```python
dfm.loc[(slice(None),['GOOGL','FB']),'open':'high']
```

Cross sections

Son más rápidos que .loc y tienen menos errores. Sin embargo sólo pueden buscar una instancia de cada nivel. (No puedes hacer slicing)

```python
dfm.xs(key = '2014-01-02', level = 0, drop_level = False)
```

```python
dfm.xs(key = 'FB', level = 'name', drop_level = False)
```

```python
dfm.xs(key = ('2014-01-02','FB'), level = ('date','name'), drop_level = False)
```

```python
dfm.xs(key = ('2014-01-02','FB'), level = (0,1), drop_level = False)
```

## Slicing multinivel sin slice()

Si el slice() es contra intuitivo y no te gusta usar comas en vez de : puedes hacer esto

```python
dfm.loc[pd.IndexSlice[1:3,'ice',:],['attack']]
```

Puedes asignar esto a una variable para hacerlo más rápido

```python
i = pd.IndexSlice 
```

```python
dfm.loc[i[1:3,'ice',:],['attack']]
```

## Moving levels around

**swaplevel**

No inplace parameter! Must me copied or reassigned

```python
dfm.swaplevel(i = 2, j = 0)
```

**reorder_levels**

```python
dfm.reorder_levels([1,0,2])
```

## Removing levels

note: both methods accept lists as parameters to remove multiple levels.

**droplevel**  
once dropped, the level is completely removed.

```python
dfm.droplevel(0)
```

**reset_index**  

Allows for level to be restored as a columns

```python
dfm.reset_index()
```

See generation columns!

```python
dfm.reset_index(level = 0, drop = False)
```

## Sorting

Sorting type level will fail because index is not sorted!

```python
dfm.loc[(1,slice('bug','water'),slice(None)),:]
```

```python
dfm.index.is_lexsorted()
```

By sorting it all gets good.

```python
dfm.sort_index(inplace = True)
```

```python
dfm.index.is_lexsorted()
```

```python
dfm.loc[(1,slice('bug','water'),slice(None)),:]
```

Customizable sorting

```python
dfm.sort_index(level = (0,1,2), ascending = [False,True,False])
```

## Flatten de columnas

```python
df = pd.read_csv('./Sets/pokemon.csv')
```

```python
dfm = df.groupby(['generation','type1']).mean()
```

```python
dfm = dfm.unstack()
```

```python
dfm
```

```python
dfm.columns = dfm.columns.to_flat_index()
```

```python
dfm
```

### O si se desea usar un sufijo
