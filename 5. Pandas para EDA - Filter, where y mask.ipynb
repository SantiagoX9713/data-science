```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Sets/pokemon.csv')
```

# Filter, where y mask

## Filter

Este dataframe tiene un montón de columnas. Una forma rápida de filtraras es con df.filter

```python
df.columns
```

Le puedes pedir un valor o una lista de valores exactas

```python
df.filter(items = ['name','is_legendary','defense'])
```

O puedes pasar un string similar a lo que buscas

```python
df.filter(like='against_')
```

Esa sería una buena forma de accesar todas esas columnas!

## Where

La idea de where es cambiar valores que no cumpan una condición


```python
df[['attack','defense']].where(lambda x: x>100, other='No pasa el límite')
```

Si no le pasas un arugmento a other los valores se vuelven NaNs automáticamente

```python
df[['attack','defense']].where(lambda x: x>100)
```

Combinado con dropna se vuelve un filtro

```python
df[['attack','defense']].where(lambda x: x>100).dropna()
```

## Mask

Igual que where pero la acción va sobre la condición positiva

```python
df[['attack','defense']].mask(lambda x: x>100, other='muy alto')
```

```python
df[['attack','defense']].mask(lambda x: x>100).dropna()
```

### np.where

El where de pandas es diferente al where de numpy  
En este caso numpy extrae los índices (si es Serie) o posiciones (si es ndarray) de los elementos que cumplan alguna condición

```python
np.where(df['attack']>100)
```

Además puede accionar sobre cierto o falso

```python
np.where(df['attack']>100,'Muy alto','Muy bajo')
```

Puede hacer cosas raras como esto:  
Donde ataque sea mayor a defensa pon el valor de ataque. Donde sea igual o menor pon el nombre del pokemon

```python
np.where(df.attack > df.defense, df.attack, df.name)
```
