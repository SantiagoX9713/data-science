# Groupby y Agg

Agg functions:
+ mean(): Compute mean of groups
+ sum(): Compute sum of group values
+ size(): Compute group sizes
+ count(): Compute count of group
+ std(): Standard deviation of groups
+ var(): Compute variance of groups
+ sem(): Standard error of the mean of groups
+ describe(): Generates descriptive statistics
+ first(): Compute first of group values
+ last(): Compute last of group values
+ nth() : Take nth value, or a subset if n is a list
+ min(): Compute min of group values
+ max(): Compute max of group values

Una agregación son varios valores reducidos a un número descriptivo de su conjunto.

## Group By Sort y Filter

```python
dfp = pd.read_csv('./Sets/pokemon.csv')
```

Aunque no aparezca type1, al ser la variable de agrupación es importante incluirla cuando se va a filtrar por grupos.

```python
gb = dfp.groupby('type1')[['attack','sp_attack','type1']]
gb
```

```python
gb.mean()
```

```python
gb.mean().sort_index()
```

```python
gb.mean().sort_values(by='attack',ascending=False)
```

Promedio del promedio de los grupos

```python
gb.mean().mean()['attack']
```

```python
media_ataque_grupos = gb.mean().mean()['attack']
```

Filtra todos los grupos cuyo promedio sea menor al promedio de todos los grupos.

Si no hubiéramos seleccionado type1 arriba no podríamos ver esa columna en el filtrado

```python
gb.filter(lambda grupo: grupo.mean()['attack']>media_ataque_grupos)
```

El primer pokemon tipo fuego tiene un ataque de 52. Esto es menos al promedio de los grupos pero pertenece a fire! Un grupo que en efecto tiene un ataque promedio mayor al promedio de los grupos.

```python
gb.filter(lambda grupo: grupo.mean()['attack']>media_ataque_grupos)['type1'].unique()
```
