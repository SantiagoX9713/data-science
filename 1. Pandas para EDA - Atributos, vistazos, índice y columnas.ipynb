## Índices y columnas

RangeIndex nos dice que cada índice es único y secuencial.

````python
df.index
````

Columns pasa una lista con las etiquetas (nombres) de las columnas

````python
df.columns
````

**set_index**

Si cambiamos índice:

````python
df.set_index('Name').head()
````

Vemos una lista de índice. (Ahora no sabemos si cada valor es único, mucho menos si esto es secuencial.)

````python
df.set_index('Name').index
````

Podemos verificar que los índices sean únicos con verify_integrity. Si la función corre, es que cada índice es único. Si no corre es que hay índices propeustos repetidos.

````python
df.set_index('Name', verify_integrity=True)
````

Si pasamos inplace = True el cambio se vuelve permanente

````python
df.set_index('Name', inplace = True)
df.head()
````

**drop**

Podemos tirar índices con drop (también requiere inplace para hacer el cambio permanente)

Drop funciona tanto para índice como para columnas sólo cambia el parámetro de index = XXXX a columns = XXXX

````python
df.drop(index = 'Braund, Mr. Owen Harris')
````

**reindex**

Podemos cambiar los índices e introducir nuevos con reindex

Tanto reindex como rename, al igual que las otras funciones que acabamos de ver, se requiere un inplace = True para hacer el cambio permanente

reindex funciona tanto para índice como para columnas sólo cambia el parámetro de index = XXXX a columns = XXXX

````python
index = ['Firefox', 'Chrome', 'Safari', 'IE10', 'Konqueror']
example = pd.DataFrame({'http_status': [200, 200, 404, 404, 301],
                  'response_time': [0.04, 0.02, 0.07, 0.08, 1.0]},
                  index=index)
example
````

Como IceWeasel y Comodo Dragon no existían previamente, estos se llenan con el valor seleccionado. Los índices ya existentes no se modifican pero cambian de lugar.

````python
new_index = ['Safari', 'Iceweasel', 'Comodo Dragon', 'IE10',
             'Chrome']
example.reindex(new_index, fill_value='Este índice es nuevo.')
````

**rename**

Podemos cambiar los nombres de los índices con un diccionario que especifique los cambios o una función

Rename funciona tanto para índice como para columnas sólo cambia el parámetro de index = XXXX a columns = XXXX

````python
example.rename(index = {'Safari':'Explorer de Apple'})
````
